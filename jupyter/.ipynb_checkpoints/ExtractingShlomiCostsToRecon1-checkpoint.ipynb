### Joining fwd and bckw

```julia
shlomi = DataFrame();
fwd = "_fwd";      # label
bkwd = "_bkwd";    # label
ids = [];          # unified id
ecs = [];          # ec number
kcats = [];         # kcats
mws = [];          # molecular weigths
sindxs = [];       #source indexes
revs = [];         # track if the image

errors = [];

shlomi[:indxs] = sindxs;
shlomi[:id] = ids;
shlomi[:ec] = ecs;
shlomi[:kcats] = kcats;
shlomi[:mws] = mws;
shlomi[:rev] = revs;


for i in 1:size(shlomisource,1)
    id = shlomisource[2][i];
    
    if endswith(id, fwd)
        id = replace(id, fwd, "");
        push!(ids, id);
        push!(ecs, shlomisource[3][i]);
        push!(kcats, [shlomisource[4][i]]);
        push!(mws, [shlomisource[5][i]]);
        push!(sindxs,[i]);
        push!(revs, 1)
        
    elseif endswith(id, bkwd)
        
        id = replace(id, bkwd, "");
        fi = find((indx) -> ids[indx] == id && revs[indx] == 1, 1:length(ids));
        if length(fi) == 0
            push!(errors,[i,id,"bkwd with missing fwd"])
        elseif length(fi) > 1
            push!(errors,[i,id,"bkwd with multiples fwd"])
        else
            push!(kcats[fi[1]], shlomisource[4][i]);
            push!(mws[fi[1]], shlomisource[5][i]);
            push!(sindxs[fi[1]], i)
        end
        
    else
        push!(ids, id);
        push!(ecs, shlomisource[3][i]);
        push!(kcats, [shlomisource[4][i]]);
        push!(mws, [shlomisource[5][i]]);
        push!(sindxs,[i]);
        push!(revs, 0);
    end
end

@assert length(ids) == length(ecs) == length(kcats) == length(mws) == length(sindxs) == length(revs)
println("errors ", length(errors));
```

```julia
r1root = "/Users/Pereiro/University/Physic/Research/Jose/Chemostat/ModelsData/Human/Recon1/Store";
recon1mat = MAT.matread(joinpath(r1root, "RECON1.mat"))["RECON1"];
```

### Matching Shlomi with Recon1

```julia
function lcs(s1::AbstractString, s2::AbstractString)
    r = ""
    i = 1
    for i in 1:length(s1)
        j = i
        while j â‰¤ length(s1) && contains(s2,s1[i:j])
            if length(r) < j - i + 1 r = s1[i:j] end
            j += 1
        end
    end
    return r
end

function ratcliff_obershelp(s1,s2,th)
    return (2*getcommom(s1,s2,0,th))/(length(s1) + length(s2));
end

function getcommom(s1,s2,commom, th)

    #lcs
    lcs_ = lcs(s1,s2);
    commom += length(lcs_);

    #Base
    if length(lcs_) < th || length(lcs_) == 0 return commom end;

    #recursion call
    s1split = remove_and_split(s1,lcs_);
    s2split = remove_and_split(s2,lcs_);
    getcommom(s1split[1],s2split[1],commom,th);
    getcommom(s1split[2],s2split[2],commom,th);

    return commom;

end

function remove_and_split(s,dl)
    r = findfirst(dl,s);
    return (s[1:first(r) - 1],s[last(r) + 1:end])
end

function formatting(id)
    id = replace(id,"__","_");
    id = replace(id,",","_");
    id = replace(id," ","");
    id = replace(id,"(","_");
    id = replace(id,")","");
    id = replace(id,"_copy1","");
    id = replace(id,"_copy2","");
    return id;
end

function lazy_delete!(col, item)
    indx = findfirst(col,item);
    col[indx] = col[end];
    pop!(col);
end
```

### Automatic extraction

```julia
r1rxns = DataFrame();
r1rxns[:id] = recon1mat["rxns"][:,1];
r1rxns[:rev] = recon1mat["rev"][:,1];
```

```julia
r1assigns = []; # the recond1 indexes already assigned
r1reminds = collect(1:length(r1rxns[:id])); # The ones still unassigned
shlomiassigns = []; # the shlomi indexes already assigned
shlominreminds = collect(1:length(shlomi[:id]));# The ones still unassigned
r1manualassigments = []; # The one assigned manually in the next cell

#Automatic
@showprogress for shi in 1:length(shlomi[:id])
    
    #shlomi id
    shid = formatting(shlomi[:id][shi]);
    
    # Track if an assigment happends
    assigned = false;
    
    #Finding perfect matchs
    for r1i in r1reminds
        
        # recon1 id
        r1id = formatting(r1rxns[:id][r1i]);
        
        # id and rev most match
        if r1id == shid && r1rxns[:rev][r1i] == shlomi[:rev][shi]
            # Assign
            push!(r1assigns, r1i);
            lazy_delete!(r1reminds, r1i)
            push!(shlomiassigns, shi);
            lazy_delete!(shlominreminds, shi)
            assigned = true;
            break;
        end 
    end
end

@assert allunique(r1assigns); 
@assert allunique(shlomiassigns);
@assert length(shlomiassigns) + length(shlominreminds) == length(shlomi[:id])
@assert length(r1assigns) + length(r1reminds) == length(r1rxns[:id])
println("Automatically assigned $(length(r1assigns)) from $(length(shlomi[:id])) shlomi ids")
```

### Manual extraction

```julia
# Manually assigning
while length(shlominreminds) > 0
    
    # Finding maxs
    shi = rand(shlominreminds);
    shid = formatting(shlomi[:id][shi]);
    shrev = shlomi[:rev][shi];
    maxeval = 0;
    maxs = [];
    for r1i in r1reminds
        r1id = formatting(r1rxns[:id][r1i]);
        eval = ratcliff_obershelp(shid, r1id, 0);
        if maxeval < eval
            push!(maxs, r1i)
            maxeval = eval;
        end
    end
    
    #Printing and selecting
    println("Assigning $(shid) :: (rev: $shrev), left $(length(shlominreminds))");
    c = 1;
    for maxi in reverse!(maxs)
        r1id = r1rxns[:id][maxi];
        r1rev = r1rxns[:rev][maxi];
        println("$c $(r1id) :: ($r1rev) ---> $(ratcliff_obershelp(shid, r1id, 0))")
        c += 1;
        c > 10 && break;
    end
    
    #Assigning
    input = readline();
    input == "q" && break;
    input == "" && continue;
    try
        
        #Parsing
        input = parse(Int, input);
        r1i = maxs[input];
        
        # Assign
        push!(r1assigns, r1i);
        push!(r1manualassigments, r1i);
        lazy_delete!(r1reminds, r1i)
        push!(shlomiassigns, shi);
        lazy_delete!(shlominreminds, shi)
        
        println("Assigned to $(r1rxns[:id][r1i]), left $(length(shlominreminds))")
        
    catch
        println("Error $input")
    end
    println();  
end

println("Assigned $(length(r1assigns)) from $(length(shlomi[:id])) shlomi ids")
```

```julia
# Writing
shlomicosts_recon1compatible_path = joinpath(storeroot, "shlomicosts_recon1compatible.csv");
shlomicosts_recon1compatible = DataFrame();
shlomicosts_recon1compatible[:r1indx] = r1assigns;
shlomicosts_recon1compatible[:r1id] = r1rxns[:id][r1assigns];
shlomicosts_recon1compatible[:shindx] = shlomiassigns;
shlomicosts_recon1compatible[:shid] = shlomi[:id][shlomiassigns];
shlomicosts_recon1compatible[:fkcat] = first.(shlomi[:kcats])[shlomiassigns];
shlomicosts_recon1compatible[:bkcat] = last.(shlomi[:kcats])[shlomiassigns];
shlomicosts_recon1compatible[:fmw] = first.(shlomi[:mws])[shlomiassigns];
shlomicosts_recon1compatible[:bmw] = last.(shlomi[:mws])[shlomiassigns];
shlomicosts_recon1compatible[:shec] = shlomi[:ec][shlomiassigns];
shlomicosts_recon1compatible[:r1rev] = r1rxns[:rev][r1assigns];
shlomicosts_recon1compatible[:shrev] = shlomi[:rev][shlomiassigns];
shlomicosts_recon1compatible[:manual] = zeros(Int,length(r1assigns));
shlomicosts_recon1compatible[:manual][findin(r1assigns, r1manualassigments)] .= 1;

CSV.write(shlomicosts_recon1compatible_path, shlomicosts_recon1compatible)
```
